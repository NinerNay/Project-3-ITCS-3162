#Regression
Regression is a supervised learning technique used to predict continuous numerical values based on one or more independent variables (features). These features influence the prediction, while the dependent variable (target) is the value we are trying to predict. Regression analysis helps us understand how changes in input variables influence the output variable.

In this project, the focus is on Linear Regression, one of the most fundamental and widely used statistical models in both machine learning and data science.

##Linear Regression

Linear regression learns from labeled datasets and models the relationship between input features and the target variable by fitting a straight line that best represents the data. The main objective of linear regression is to find the line that minimizes the difference (error) between the predicted values and the actual data points.

This relationship is represented by the equation:
* y = mx +b

Where:
- y = predicted value (dependent variable)
- x = input feature (independent variable)
- m = slope of the lines (how much y changes when x changes)
- b = the intercept (value of y when x = 0)

In practice, linear regression identifies the optimal values of
m and b that minimize the Mean Squared Error (MSE) between predicted and actual outcomes. This process enables the model to generalize patterns from the training data and make accurate predictions on new, unseen data.

##Pre-Processing
After gaining an initial understanding of the dataset, I began the preprocessing stage by addressing missing values. For numerical features, missing entries were replaced with the mean of each column to preserve the overall distribution of the data. For categorical features, missing values were filled with the label "None" to indicate the absence of a category rather than treating them as missing data.

Next, I converted all categorical variables into numerical format to make the dataset compatible with the Linear Regression model, which requires numerical input. At this stage, I chose to perform minimal preprocessing to observe how the raw dataset would perform with a basic model. I initially focused on numerical features only to simplify the first experiment and avoid potential complications related to categorical encoding.

##Modeling
For this experiment, I used the training dataset provided by Kaggle, which was already divided into separate training and testing files. I began with a basic Linear Regression model to observe how well the model could perform with minimal preprocessing. This initial approach provided a simple yet effective starting point for understanding the predictive power of the dataset before introducing more complex modeling techniques.

##Evaluation

The Linear Regression model achieved an RMSE (Root Mean Squared Error) score of approximately 0.2168 and an R score of about 0.7480. These results suggest that the model explains roughly 75% of the variance in housing prices, which is a reasonable outcome given the limited preprocessing performed.

The RMSE indicates how far, on average, the models predictions deviate from the actual house prices  lower values represent better performance. To further assess model quality, I examined the residuals, which are the differences between the actual and predicted sale prices. Ideally, residuals should be randomly distributed around zero, showing no clear pattern.

In this case, the residual plot showed that most residuals were centered around zero, suggesting that the Linear Regression model captured the main trends in the data reasonably well. However, there is still room for improvement, as some variation remains unexplained, and additional preprocessing or model adjustments could further enhance performance.

#Experiment 2

For the second experiment, I introduced additional preprocessing steps to improve model performance. Specifically, I standardized the numerical features using the StandardScaler function from scikit-learn. This transformation adjusts each numerical feature to have a mean of 0 and a standard deviation of 1, ensuring that all features contribute equally to the model regardless of their original scales.

In addition, I used dummy variables to properly encode the categorical features. This step helps the model interpret categorical information numerically without introducing bias from relationships that do not actually exist.

After scaling and encoding the data, I retrained the Linear Regression model. The performance improved, achieving an RMSE score of 0.1737 and an R score of 0.8382. These results indicate that the model was able to explain approximately 84% of the variance in housing prices, marking a clear improvement over the first experiment.

The residual plot also showed a more even distribution of residuals around zero, suggesting that the scaled model handles the data more effectively. However, while the overall performance increased, there is still a possibility of overfitting, meaning the model may perform well on the training data but struggle to generalize to unseen data. This insight sets the stage for further refinement in the next experiment.

#Experiment 3
In this experiment, I decided to change the model entirely by introducing Ridge and Lasso Regression models. These models apply a technique known as regularization, which helps reduce overfitting by adding a penalty term to the models cost function. Regularization discourages the model from assigning excessively large weights to any one feature, promoting better generalization to unseen data.

- Ridge Regression uses L2 regularization, which adds the squared magnitude of the coefficients as a penalty term.

- Lasso Regression uses L1 regularization, which adds the absolute value of the coefficients as a penalty and can also drive some coefficients to zero, effectively performing feature selection.

#Impact
Predictive modeling in the housing market can have several positive impacts. By accurately estimating home values, homeowners, buyers, and real estate professionals can make more informed financial decisions. These models can assist in determining fair market value, help sellers price their homes competitively, and support tax assessments and urban planning.

However, predictive housing models also raise important ethical and social concerns. One major issue is neighborhood bias  models may unintentionally favor certain areas if the training data reflects existing economic or social inequalities. As a result, homes in historically wealthy neighborhoods may consistently be predicted at higher values, while those in underrepresented or lower-income areas may be undervalued. This can contribute to economic and racial inequities in the housing market. Therefore, while predictive housing models can be highly useful, they must be developed and applied with fairness, transparency, and social awareness in mind.
